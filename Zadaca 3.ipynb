2. a) Find the mean, median and average scores in mathematics, reading and writing. <br>
   b) Find the mean, median and average scores for males and females.<br>
   c) What is the coefficient of variation of the scores? <br>

```python
# a) finding mean and median for math, reading and writing scores
mean_math = X['MathScore'].mean()
median_math = X['MathScore'].median()

mean_reading = X['ReadingScore'].mean()
median_reading = X['ReadingScore'].median()

mean_writing = X['WritingScore'].mean()
median_writing = X['WritingScore'].median()

print (f' For math scores mean is {mean_math} and median is {median_math}.')
print (f' For reading scores mean is {mean_reading} and median is {median_reading}.')
print (f' For writing scores mean is {mean_writing} and median is {median_writing}.')
```

```python
#replacing entries in Gender with 0 and 1 so I would be able to calculate measures of average later
X['Gender'] = X['Gender'].replace({'male': 0, 'female': 1})
X
```

```python
#b) finding mean and median for males and females
gender_scores = X.groupby('Gender').agg({
    'MathScore': ['mean', 'median'],
    'ReadingScore': ['mean', 'median'],
    'WritingScore': ['mean', 'median']
}).rename(columns={'mean': 'Mean', 'median': 'Median'})

print (gender_scores)
```

```python
# c) calculating coefficient of variation of the scores in math, reading and writing
coef_math = X['MathScore'].std() / X.MathScore.mean() * 100
coef_reading = X['ReadingScore'].std() / X.ReadingScore.mean() * 100
coef_writing = X['WritingScore'].std() / X.WritingScore.mean() * 100

print (f'For math scores coefficient of variation is {coef_math:.2f}.')
print (f'For reading scores coefficient of variation is {coef_reading:.2f}.')
print (f'For writing scores coefficient of variation is {coef_writing:.2f}.')
```

3. a) Use a statistical test to test the hypothesis:<br>
H0: the mean mathematical scores of females and males are equal.



```python
from scipy import stats
```

```python
# using t-test to see if there are differences between two groups (assuming the basic assumptions of the t-test like random sampling or normal distribution are satisfied)
males = X.query('Gender == 0')
females = X.query('Gender == 1')
stats.ttest_ind(males['MathScore'],females['MathScore'])

# p<0.05 meaning there is a statistically significant difference between groups, we can reject the null hypothesis
```

4. a) Use a statistical test to test the hypothesis:<br>
   H0 Gender did not affect the writing scores.<br>
   H1 Gender affected the writing score.<br>

```python
#using t-test to see if there are differences in writing scores based on gender
males = X.query('Gender == 0')
females = X.query('Gender == 1')
stats.ttest_ind(males['WritingScore'],females['WritingScore'])

# p<0.05 meaning there is a statistically significant difference between groups, we can reject the null hypothesis and confirm H1
```

5. a) Statisticaly test the distribuitions of female and male scores across the three scores columns:<br>
   b) Plot the distributions for each of the separate scores in both genders<br>


```python
from scipy.stats import shapiro
# extracting scores in three columns for males and females
scores_male = X[X['Gender'] == 0][['MathScore', 'ReadingScore', 'WritingScore']]
scores_female = X[X['Gender'] == 1][['MathScore', 'ReadingScore', 'WritingScore']]

# a) using Shapiro-Wilk test to see are the distributions normal

for score_column in ['MathScore', 'ReadingScore', 'WritingScore']:
    stat, p_value_male = shapiro(scores_male[score_column])
    stat, p_value_female = shapiro(scores_female[score_column])

    print(f"{score_column} for Males: Shapiro-Wilk Test is {stat}, P-value = {p_value_male}")
    print(f"{score_column} for Females: Shapiro-Wilk Test is {stat}, P-value = {p_value_female}")
    print("\n")
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 0 is male, 1 is female

# making 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# b) Plot histograms for each score column by gender
score_columns = ['MathScore', 'ReadingScore', 'WritingScore']
for i in range(len(score_columns)):
    sns.histplot(data=X, x=score_columns[i], hue='Gender', kde=True, bins=30, ax=axes[i])
    axes[i].set_title(f'Distribution of {score_columns[i]} Scores by Gender')

plt.tight_layout()
plt.show()
```
