8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
X_tmp = X.copy()
# a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.
X_tmp.loc[X_tmp.EthnicGroup.isna(),'EthnicGroup'] = X_tmp.loc[:,'EthnicGroup'].mode()
X_tmp.loc[X_tmp.ParentMaritalStatus.isna(),'ParentMaritalStatus'] = X_tmp.loc[:,'ParentMaritalStatus'].mode()
X_tmp.loc[X_tmp.WklyStudyHours.isna(),'WklyStudyHours'] = X_tmp.loc[:,'WklyStudyHours'].mode()

#b) For the feature ParentEduc replace NaN value with a string no formal education.
X_tmp["ParentEduc"].fillna("no formal educaiton", inplace=True)

#c) For the feature TestPrep replace NaN value with a string none.
X_tmp["TestPrep"].fillna("none", inplace=True)

#d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values.
X_tmp.dropna(subset=['PracticeSport', 'IsFirstChild'], inplace=True)
X_tmp.reset_index(drop=True, inplace=True)

#e) For the feature NrSiblings replace the NaN value with the median value.
mean_value=X_tmp.NrSiblings.median()
X_tmp["NrSiblings"].fillna(mean_value, inplace=True)

#f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.
X_tmp["TransportMeans"].fillna("walking", inplace=True)

X_tmp
```

10. The final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
encoder = OneHotEncoder(handle_unknown="ignore")
encoder.fit(X_tmp[['Gender', 'LunchType']])
X_novi = encoder.transform(X_tmp[['Gender', 'LunchType']])
X_novi = pd.DataFrame(X_novi.toarray(), columns=encoder.get_feature_names_out())
X_novi
```

```python
X_tmp = pd.concat((X_tmp, X_novi), axis=1)
X_tmp
```

```python
columns_to_labelencode=['EthnicGroup', 'ParentEduc', 'TestPrep', 'ParentMaritalStatus','PracticeSport', 'IsFirstChild', 'TransportMeans', 'WklyStudyHours']
label_encoder = LabelEncoder()
for column in columns_to_labelencode:
    X_tmp[column] = label_encoder.fit_transform(X_tmp[column])
print(X_tmp.info())

X_tmp
```
